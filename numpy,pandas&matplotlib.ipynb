```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, y1, color='blue', label='Sin')
ax.plot(x, y2, color='red', label='Cos')

ax.set_title("Perbandingan Sin & Cos")
ax.legend()
plt.show()
```

```python
# Program to calculate average marks of students

num_students = int(input("Enter the number of students: "))

total_marks = 0

for i in range(1, num_students + 1):
    marks = float(input(f"Enter marks for student {i}: "))
    total_marks += marks

average = total_marks / num_students

print(f"\nTotal Marks: {total_marks}")
print(f"Average Marks: {average:.2f}")
```

## Kalkulator Indeks Jisim Badan (BMI) Interaktif

Masukkan berat badan anda dalam kilogram (kg) dan tinggi anda dalam meter (m) menggunakan slider di bawah untuk mengira BMI anda secara langsung.

```python
import ipywidgets as widgets
from IPython.display import display, HTML
import matplotlib.pyplot as plt
import numpy as np

# Widgets untuk input berat dan tinggi
weight_slider = widgets.FloatSlider(
    value=70, # Nilai permulaan
    min=30,   # Berat minimum
    max=150,  # Berat maksimum
    step=0.1, # Langkah perubahan
    description='Berat (kg):',
    continuous_update=True,
    orientation='horizontal',
    readout=True,
    readout_format='.1f',
)

height_slider = widgets.FloatSlider(
    value=1.70, # Nilai permulaan
    min=1.00,   # Tinggi minimum
    max=2.20,   # Tinggi maksimum
    step=0.01,  # Langkah perubahan
    description='Tinggi (m):',
    continuous_update=True,
    orientation='horizontal',
    readout=True,
    readout_format='.2f',
)

# Paparkan slider
display(weight_slider, height_slider)

def calculate_bmi(weight_kg, height_m):
    """Mengira BMI dan mengkategorikan berdasarkan berat dan tinggi."""
    if height_m <= 0:
        return 0, 'Tidak Sah', 'black', 0, 0

    bmi = weight_kg / (height_m ** 2)
    category = ''
    color = ''

    # Menentukan kategori BMI dan warna
    if bmi < 18.5:
        category = 'Kurang Berat'
        color = 'orange'
    elif 18.5 <= bmi < 25:
        category = 'Berat Normal'
        color = 'green'
    elif 25 <= bmi < 30:
        category = 'Lebih Berat'
        color = 'yellow'
    else:
        category = 'Obes'
        color = 'red'

    # Kira julat berat ideal untuk tinggi yang diberikan (BMI 18.5 - 24.9)
    min_ideal_weight = 18.5 * (height_m ** 2)
    max_ideal_weight = 24.9 * (height_m ** 2)

    return bmi, category, color, min_ideal_weight, max_ideal_weight



def display_bmi_results(weight_kg, height_m):
    """Memaparkan hasil BMI, kategori, dan visualisasi."""
    bmi, category, color, min_ideal_weight, max_ideal_weight = calculate_bmi(weight_kg, height_m)

    # Paparan Data Terperinci
    display(HTML(f"<h3>Keputusan BMI Anda:</h3>"))
    display(HTML(f"<p><b>Berat Badan:</b> {weight_kg:.1f} kg</p>"))
    display(HTML(f"<p><b>Tinggi:</b> {height_m:.2f} m</p>"))
    display(HTML(f"<p><b>Indeks Jisim Badan (BMI) Anda:</b> <span style='color:{color};'><b>{bmi:.2f}</b></span></p>"))
    display(HTML(f"<p><b>Kategori BMI:</b> <span style='color:{color};'><b>{category}</b></span></p>"))
    if category != 'Tidak Sah':
      display(HTML(f"<p><b>Julat Berat Badan Sihat (untuk tinggi ini):</b> {min_ideal_weight:.1f} kg - {max_ideal_weight:.1f} kg</p>"))

    # Visualisasi BMI
    fig, ax = plt.subplots(figsize=(10, 2))

    # Definisi julat BMI dan warna
    bmi_ranges = [0, 18.5, 25, 30, 35, 40] # Tambah julat atas untuk visualisasi
    category_labels = ['Kurang Berat', 'Normal', 'Lebih Berat', 'Obes (Kelas I)', 'Obes (Kelas II+)']
    category_colors = ['orange', 'green', 'yellow', '#FF7F50', 'red'] # Warna berbeza untuk Obes

    # Lukiskan bar kategori BMI
    for i in range(len(bmi_ranges) - 1):
        ax.axvspan(bmi_ranges[i], bmi_ranges[i+1], facecolor=category_colors[i], alpha=0.7)

    # Tandakan BMI pengguna
    ax.axvline(bmi, color='black', linestyle='--', linewidth=2, label=f'BMI Anda ({bmi:.2f})')

    ax.set_yticks([]) # Sembunyikan tick y-axis
    ax.set_xticks(bmi_ranges)
    ax.set_xticklabels([f'{r:.1f}' for r in bmi_ranges])
    ax.set_xlim(0, max(40, bmi + 5)) # Pastikan julat x mencukupi
    ax.set_title('Skala Kategori BMI Anda', fontsize=14)
    ax.legend()
    ax.grid(False)

    # Tambah label kategori
    for i in range(len(bmi_ranges) - 1):
        mid_point = (bmi_ranges[i] + bmi_ranges[i+1]) / 2
        if mid_point < max(40, bmi + 5): # Hanya label jika dalam julat xlim
            ax.text(mid_point, 0.5, category_labels[i], ha='center', va='center', color='black', fontsize=9, rotation=0)

    plt.show()


    # Sambungkan slider dengan fungsi paparan
interactive_plot = widgets.interactive(display_bmi_results, weight_kg=weight_slider, height_m=height_slider)

# Paparkan output interaktif
display(interactive_plot)
```
